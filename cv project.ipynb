### 1. Identifying Inputs and Output

Output is Label column and remaining columns are the inputs in the dataset

### 2. EDA (Exploratory Data Analysis)

```python
sns.countplot(data = df, x = "Label")
plt.show()
```

#### From this plot we can conclude that the data is slightly imbalanced to accuracy is not a good metric to check the model's performance.

### 3. Segregrate inputs and output

```python
X = df.drop("Label", axis = 1)

y = df["Label"]
```

### 4. Split the data into train and test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
```

### 5. Doing Data Preparation on trained data

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

X_train_transformed = pd.DataFrame(std_scaler.fit_transform(X_train),
                                   columns = std_scaler.get_feature_names_out(),
                                   index = X_train.index)
```

### 6. Doing Data Preparation on test data

```python
X_test_transformed = pd.DataFrame(std_scaler.transform(X_test),
                                 columns = std_scaler.get_feature_names_out(),
                                 index = X_test.index)
```

### 7. Building model's on Processed train data

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier(min_samples_leaf = 2)

%time DT.fit(X_train_transformed, y_train)
```

```python
%time y_test_pred = DT.predict(X_test_transformed)
```

```python
from sklearn import metrics

metrics.accuracy_score(y_test, y_test_pred)
```

```python
print(metrics.classification_report(y_test, y_test_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(min_samples_leaf = 2)

%time RF.fit(X_train_transformed, y_train)
```

```python
%time y_test_pred = RF.predict(X_test_transformed)
```

```python
print(metrics.accuracy_score(y_test, y_test_pred))
```

```python
print(metrics.classification_report(y_test, y_test_pred))
```

```python
from sklearn.naive_bayes import GaussianNB

GB = GaussianNB()

%time GB.fit(X_train_transformed, y_train)
```

```python
%time y_test_pred = GB.predict(X_test_transformed)
```

```python
metrics.accuracy_score(y_test, y_test_pred)
```

```python
print(metrics.classification_report(y_test, y_test_pred))
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()

%time LR.fit(X_train_transformed, y_train)
```

```python
%time y_test_pred = LR.predict(X_test_transformed)
```

```python
metrics.accuracy_score(y_test, y_test_pred)
```

```python
print(metrics.classification_report(y_test, y_test_pred))
```
